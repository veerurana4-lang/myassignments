```python
'''Q1. What are the characteristics of tuples? Is tuple immutable?'''
#ans:
'''Tuples are ordered collection
Tuples are defined in parantheses ()
Tuples can store different data types
Yes Tuples are immutable'''
tup = (1,2,3,4,"ajay",1.2,3+5j,4,4)
```

```python
'''Q2. What are the two tuple methods in python? Give an example of each method. Give a reason why
tuples have only two in-built methods as compared to Lists.'''
#ans:
#count()
result = tup.count(4)

ind = tup.index(3)
print(result)
print(ind)

# output : 3,2
#we can't modify tuples and we can't use append(),exten(),pop() so tuples have only two methods
```

```python
'''Q3. Which collection datatypes in python do not allow duplicate items? Write a code using a set to remove
duplicates from the given list.'''
#ans:
'''Set is a collection datatype which do not allow duplicate values it gives only unique values'''
lis = [1, 1, 1, 2, 1, 3, 1, 4, 2, 1, 2, 2, 2, 3, 2, 4, 3, 1, 3, 2, 3, 3, 3, 4, 4, 1, 4, 2, 4, 3, 4, 4]
p = set(lis)
print(p)

#output : {1, 2, 3, 4}
```

```python
'''Q4. Explain the difference between the union() and update() methods for a set. Give an example of
each method.'''
#ans:
set1 = {1, 2, 3, 4}
set2 = {5,6}
result = set1.union(set2) #union makes a new set 
print(result)
print(set1)
#output : {1, 2, 3, 4, 5, 6}
#{1, 2, 3, 4}

set1.update(set2) # update is used where i need to modify in a data 
print(set1)
#output : {1, 2, 3, 4, 5, 6}
```

```python
'''Q5. What is a dictionary? Give an example. Also, state whether a dictionary is ordered or unordered.'''
#ans : 
'''Dictionary is a collection of key,value pairs. keys are unique '''
student = {
    "name": "raj",
    "age": 22,
    "course": "Data Science"}

print(student)
#Dictionary is ordered (Python 3.7+)
```

```python
'''Q6. Can we create a nested dictionary? If so, please give an example by creating a simple one-level
nested dictionary.'''
#ans:
'''yes we can create a nested dictionary'''
student = {"name":"ravi", "height":155,"weight":55,"education":{"branch":"mech","level":"graduation"}}
print(student)
```

```python
'''Q7. Using setdefault() method, create key named topics in the given dictionary and also add the value of
the key as this list ['Python', 'Machine Learning’, 'Deep Learning']'''
#ans:
dict1 = {'language' : 'Python', 'course': 'Data Science Masters'}
dict1.setdefault('topics', ['Machine Learning', 'Deep Learning'])
dict1

'''output :
 {'language': 'Python',
 'course': 'Data Science Masters',
 'topics': ['Machine Learning', 'Deep Learning']}'''
```
